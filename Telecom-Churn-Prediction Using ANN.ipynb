```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df=pd.read_csv("Tel_Churn_Modelling.csv")
df.head()
```

```python
df.drop('customerID',axis=1,inplace=True)
df
```

```python
df=df[df.TotalCharges!=' ']
df.shape
```

```python
df.TotalCharges=pd.to_numeric(df.TotalCharges)
for col in df:
    print(f'{col} : {df[col].unique()}')
```

```python
df.replace({'No internet service':'No'},inplace=True)
df.replace({'No phone service':'No'},inplace=True)
df.replace({'Male':1,'Female':0},inplace=True)
for col in df:
    print(f'{col} : {df[col].unique()}')
```

```python
df.replace({'Yes':1,'No':0},inplace=True)
for col in df:
    print(f'{col} : {df[col].unique()}')
```

```python
df=pd.get_dummies(data=df,columns=['InternetService','Contract','PaymentMethod'])
```

```python
for col in df:
    print(f'{col} : {df[col].unique()}')
```

```python
col_to_scale=['tenure','MonthlyCharges','TotalCharges']
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
df[col_to_scale]=sc.fit_transform(df[col_to_scale])
```

```python
df
```

```python
for col in df:
    print(f'{col} : {df[col].unique()}')
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(df.drop('Churn',axis=1), df['Churn'], test_size=0.2, random_state=0)
X_train.shape, X_test.shape
```

```python
import tensorflow as tf
from tensorflow import keras

model = keras.Sequential([
    keras.layers.Dense(20, input_shape=(26,), activation = 'relu'),
    keras.layers.Dense(15, activation = 'relu'),
    keras.layers.Dense(1, activation = 'sigmoid')
])

model.compile( optimizer='adam',
               loss='binary_crossentropy',
               metrics=['accuracy']
             )
```

```python
model.fit(X_train,y_train,epochs=100)
```

```python
y_pred = model.predict(X_test)
y_pred[:10]
```

```python
y_p=[]
for element in y_pred:
    if element > 0.5:
        y_p.append(1)
    else:
        y_p.append(0)
        
y_p[:10],y_test[:10]
```

```python
print("Confusion Matrix :\n",pd.crosstab(y_test, np.array(y_p), rownames=['Actual'], colnames=['Predicted'], margins = True))
```

```python
from sklearn.metrics import confusion_matrix, classification_report
print(classification_report(y_test,y_p))
```
